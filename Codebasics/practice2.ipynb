```python
import csv, statistics
```

```python
def calculate_rating_stats(data, industry = None):
    rating = []
    for row in data:
        if row[3]!='NULL' and (not industry or row[1] == industry):
            rating.append(float(row[3]))
    max_rating = max(rating)
    min_rating = min(rating)
    mean_rating = statistics.mean(rating)
    return max_rating, min_rating, mean_rating
```

```python
with open("movies.csv") as f:
    data = list(csv.reader(f))
    header = data[0]
    data = data[1:]
max_rating, min_rating, mean_rating = calculate_rating_stats(data)
print(f"All records: Max rating = {max_rating}, Min rating = {min_rating}, and Avg rating = {mean_rating}. \n")

max_rating, min_rating, mean_rating = calculate_rating_stats(data, industry = "Bollywood")
print(f"Bollywood : Max rating = {max_rating}, Min rating = {min_rating}, and Avg rating = {mean_rating}. \n")

max_rating, min_rating, mean_rating = calculate_rating_stats(data, industry = "Hollywood")
print(f"Hollywood : Max rating = {max_rating}, Min rating = {min_rating}, and Avg rating = {mean_rating}.")
```
